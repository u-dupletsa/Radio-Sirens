## N-body maps

```python
ska_color_pink = '#d60068'
ska_color_intermediate = '#840068'
ska_color_purple = '#180068'
radio_color = '#f5e15f'
```

```python
nside = 16
npix = hp.nside2npix(nside)
lmax = 3*nside 

# redhsift range for reference
zmin = 0.005
zmax = 3.025

H0_value = 67.7
# Reference cosmology
cosmo_ref = icarogw.cosmology.astropycosmology(zmax=20.)
cosmo_ref.build_cosmology(FlatLambdaCDM(H0=H0_value,Om0=0.308)) 
```

```python
PATH_TO_DENSITY_MATRIX = '../../hi_data/'
density_matrix = pd.read_hdf(PATH_TO_DENSITY_MATRIX + 'map4icaro_lmax' + str(lmax) + '_zmin' + str(zmin) + '_zmax' + str(zmax) + '_nside' + str(nside) + '.hdf5').to_numpy()
```

```python
redshift_grid = np.array([0.005, 0.015, 0.035, 0.06 , 0.085, 0.11 , 0.135, 0.17 , 0.205,
        0.24 , 0.28 , 0.32 , 0.36 , 0.405, 0.455, 0.505, 0.56 , 0.62 ,
        0.68 , 0.745, 0.82 , 0.895, 0.975, 1.06 , 1.155, 1.26 , 1.365,
        1.485, 1.615, 1.755, 1.905, 2.05 , 2.205, 2.39 , 2.585, 2.795,
        3.025])
pixel_grid = np.arange(0, npix, 1).astype(int) # generates an array of (integer) numbers from 0 to npix-1
HI = icarogw.HI.HI_map(redshift_grid,pixel_grid,density_matrix, bGW=1, alphaGW=0)
HI_bias2 = icarogw.HI.HI_map(redshift_grid,pixel_grid,density_matrix, bGW=2, alphaGW=0)
HI_bias05 = icarogw.HI.HI_map(redshift_grid,pixel_grid,density_matrix, bGW=2, alphaGW=2)
```

```python
from scipy.ndimage import gaussian_filter1d
gaussian_sigma = 6.5
```

```python
fig = plt.figure(figsize=(6,4))
my_linewidth = 2.5
pixx = 762 # pick a pixel number (in 0 to npix-1) to plot the density of hydrogen as function of redshift
ra  = 150   # degrees
dec = 30     # degrees
z_plot = np.linspace(0.005,10,1000)
pixels_array = pixx*np.ones_like(z_plot)
gc = HI.drho_dzdomega(z_plot,pixels_array,cosmo_ref)
# smooth the data points
gc_smooth = gaussian_filter1d(gc, sigma=gaussian_sigma)
plt.semilogx(z_plot,gc_smooth,label=r'\rm $b_{\rm GW} = 1$, $\alpha_{\rm GW} = 0$', linewidth=my_linewidth, color=ska_color_purple, linestyle='solid')

gc = HI_bias2.drho_dzdomega(z_plot,pixels_array,cosmo_ref)
gc_smooth = gaussian_filter1d(gc, sigma=gaussian_sigma)
plt.semilogx(z_plot,gc_smooth,label=r'\rm $b_{\rm GW} = 2$, $\alpha_{\rm GW} = 0$', linewidth=my_linewidth, color=ska_color_purple, linestyle='dashed')

gc = HI_bias05.drho_dzdomega(z_plot,pixels_array,cosmo_ref)
gc_smooth = gaussian_filter1d(gc, sigma=gaussian_sigma)
plt.semilogx(z_plot,gc_smooth,label=r'\rm $b_{\rm GW} = 2$, $\alpha_{\rm GW} = 2$', linewidth=my_linewidth, color=ska_color_purple, linestyle='dotted')
plt.xscale('log')

plt.xlim(0.005, 10)
plt.xlabel(r'\rm Redshift $z$', fontsize =16)
plt.ylabel(r'$\rho_{\rm BH} / \bar{\rho}_{\rm BH}$', fontsize =16)
plt.legend(fontsize=14, frameon=False)
plt.title(r'\rm RA = {:.0f}$^{{\circ}}$, Dec = {:.0f}$^{{\circ}}$'.format(ra, dec), fontsize =16)
plt.savefig('HI_density_profile_smooth_bias.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
