```python
import warnings
warnings.filterwarnings("ignore", "Wswiglal-redir-stdio")

import icarogw
import healpy as hp
import numpy as np
import pandas as pd
from  astropy.cosmology import FlatLambdaCDM

import bilby
```

```python
Ninjections = int(1e7)  # use a large number
np.random.seed(42) # fix for reproducibility
dl_prior_injections = bilby.gw.prior.UniformSourceFrame(name='luminosity_distance', minimum=5,maximum=40000)
# create dataframe of injections for snr calculation
mass_1_inj = np.random.uniform(5,300,size=Ninjections)
mass_2_inj = np.random.uniform(5,300,size=Ninjections)
dl_inj = dl_prior_injections.sample(Ninjections)
theta_jn_inj = np.arccos(np.random.uniform(-1,1,size=Ninjections))
polarization_inj = np.random.uniform(0,np.pi,size=Ninjections)
phase_inj = np.random.uniform(0,2*np.pi,size=Ninjections)
a_1_inj = np.random.uniform(0.005,0.99,size=Ninjections)
a_2_inj = np.random.uniform(0.005,0.99,size=Ninjections)
tilt_1_inj = np.arccos(np.random.uniform(-1,1,size=Ninjections))
tilt_2_inj = np.arccos(np.random.uniform(-1,1,size=Ninjections))
phi_12_inj = np.random.uniform(0,2*np.pi,size=Ninjections)
phi_jl_inj = np.random.uniform(0,2*np.pi,size=Ninjections)
times_inj = np.zeros(Ninjections) # coalescence time at the geocenter


# create dataframe with events to pass to GWFish for SNR computation
        
data_inj = {'mass_1': mass_1_inj,
        'mass_2': mass_2_inj,
        'luminosity_distance': dl_inj,
        'theta_jn': theta_jn_inj,
        'ra': np.random.uniform(0,2*np.pi,size=Ninjections),
        'dec': np.arccos(np.random.uniform(-1,1,size=Ninjections))-np.pi/2,
        'psi': polarization_inj,
        'phase': phase_inj,
        'geocent_time': times_inj,
        'a_1': a_1_inj,
        'a_2': a_2_inj,
        'tilt_1': tilt_1_inj,
        'tilt_2': tilt_2_inj,       
        'phi_12': phi_12_inj,
        'phi_jl': phi_jl_inj,
        'inj_prior': dl_prior_injections.prob(dl_inj)}
df_inj = pd.DataFrame(data_inj)
```

```python
df_inj.to_hdf('gw_data/bbh_1e7_injections_for_gwfish_5_300_Msol.hdf5', key='df', mode='w')
```
